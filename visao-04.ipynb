```python
# Importação das bibliotecas
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Carregue a imagem (substitua 'path_to_your_image.jpg' pelo caminho da sua imagem)
image_path = '/content/drive/MyDrive/Pasta sem nome/Pasta sem nome/amsler-II-1.png'  # Substitua pelo caminho da sua imagem
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Verifique se a imagem foi carregada corretamente
if image is None:
    raise ValueError("A imagem não foi carregada. Verifique o caminho do arquivo.")

# Defina uma função de correção radiométrica
def radiometric_correction(image, a=1.0, b=0.0):
    """
    Corrige a distorção radiométrica com uma função linear.
    :param image: Imagem em escala de cinza.
    :param a: Coeficiente de ganho.
    :param b: Termo de deslocamento.
    :return: Imagem corrigida.
    """
    corrected_image = a * image + b
    corrected_image = np.clip(corrected_image, 0, 255)  # Certifique-se de que os valores estejam no intervalo [0, 255]
    return corrected_image.astype(np.uint8)

# Aplique a correção radiométrica
a = 1.2  # Coeficiente de ganho
b = -30  # Termo de deslocamento
corrected_image = radiometric_correction(image, a, b)

# Mostre a imagem original e a imagem corrigida
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Imagem Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagem Corrigida')
plt.imshow(corrected_image, cmap='gray')
plt.axis('off')

plt.show()
```
